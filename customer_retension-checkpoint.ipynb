# Exploratory Data Analysis

```python
#Printing their object type and their unique values
for column in ren:
    if ren[column].dtype==object:
        print(str(column)+' : '+str(ren[column].unique()))
        print(ren[column].value_counts())
        print("********************************************************************")
        print('\n')
```

# Data Visiulization

```python
#loading gender count
sns.countplot(x="1Gender of respondent",data=ren,palette='Spectral')
plt.title("Number of Male and Female")
plt.show()
print(ren["1Gender of respondent"].value_counts())
```

```python
#In the above chart we can see female is 181 and male is 88.
```

```python
#Loading which city do you shop online from?
sns.countplot(x="3 Which city do you shop online from?",data=ren,palette='cool')
plt.title("which city do you shop from")
plt.show()
print(ren["3 Which city do you shop online from?"].value_counts())
```
